```python
%pylab inline
import numpy as np
```

```python
full_results = np.load('./full_POD_results/results.npy', allow_pickle=True).item()
midEps_results = np.load('./midEps_POD_results/results.npy', allow_pickle=True).item()
mid_results = np.load('./mid_POD_results/results.npy', allow_pickle=True).item()
close_results = np.load('./close_POD_results/results.npy', allow_pickle=True).item()
bdry_results = np.load('./bdry_POD_results/results.npy', allow_pickle=True).item()
```

```python
transient = 100 # skip first transient timesteps
n_runs = len(full_results['Cd_true'])

Cd_keys = ['Cd_true','Cd_10','Cd_25','Cd_100','Cd_500']
Cl_keys = ['Cl_true','Cl_10','Cl_25','Cl_100','Cl_500']

basic_statistics = {}

basic_statistics['mean_Cd_full'] = [[np.mean(full_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['mean_Cd_midEps'] = [[np.mean(midEps_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['mean_Cd_mid'] = [[np.mean(mid_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['mean_Cd_close'] = [[np.mean(close_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['mean_Cd_bdry'] = [[np.mean(bdry_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]

basic_statistics['mean_Cl_full'] = [[np.mean(full_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['mean_Cl_midEps'] = [[np.mean(midEps_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['mean_Cl_mid'] = [[np.mean(mid_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['mean_Cl_close'] = [[np.mean(close_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['mean_Cl_bdry'] = [[np.mean(bdry_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]

basic_statistics['std_Cd_full'] = [[np.std(full_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['std_Cd_midEps'] = [[np.std(midEps_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['std_Cd_mid'] = [[np.std(mid_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['std_Cd_close'] = [[np.std(close_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]
basic_statistics['std_Cd_bdry'] = [[np.std(bdry_results[key][j][transient:]) for j in range(n_runs)] for key in Cd_keys]

basic_statistics['std_Cl_full'] = [[np.std(full_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['std_Cl_midEps'] = [[np.std(midEps_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['std_Cl_mid'] = [[np.std(mid_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['std_Cl_close'] = [[np.std(close_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
basic_statistics['std_Cl_bdry'] = [[np.std(bdry_results[key][j][transient:]) for j in range(n_runs)] for key in Cl_keys]
```

```python
figure(figsize = (15,12))
label_fs = 14

figure_stats = [
    ['mean_Cd_full','mean_Cd_midEps','mean_Cd_mid','mean_Cd_close','mean_Cd_bdry'],
    ['mean_Cl_full','mean_Cl_midEps','mean_Cl_mid','mean_Cl_close','mean_Cl_bdry'],
    ['std_Cd_full','std_Cd_midEps','std_Cd_mid','std_Cd_close','std_Cd_bdry'],
    ['std_Cl_full','std_Cl_midEps','std_Cl_mid','std_Cl_close','std_Cl_bdry']
]

for j in range(4):
    for i in range(5):
        subplot(4,5,5*j+i+1)
        
        for k in [3,4]:
            plot(basic_statistics[figure_stats[j][i]][0], 
                 basic_statistics[figure_stats[j][i]][k], 
                 '.', markersize=10)
            
        qmin = np.min(np.concatenate(basic_statistics[figure_stats[j][i]]))
        qmax = np.max(np.concatenate(basic_statistics[figure_stats[j][i]]))
        xlim([qmin - 0.05*(qmax-qmin), qmax + 0.05*(qmax-qmin)])
        ylim([qmin - 0.05*(qmax-qmin), qmax + 0.05*(qmax-qmin)])
        plot([-50,50],[-50,50],'k--', linewidth=0.5, zorder=-1)
        
        if j == 0:
            if i == 0: title('Full domain POD', fontsize = label_fs)
            elif i == 1: title('Mid + epsilon domain POD', fontsize = label_fs)
            elif i == 2: title('Mid domain POD', fontsize = label_fs)
            elif i == 3: title('Close domain POD', fontsize = label_fs)
            else: title('Boundary POD', fontsize = label_fs)
                
        if i == 0:
            if j == 0: ylabel(r'Pred Mean $C_d$', fontsize = label_fs)
            elif j == 1: ylabel(r'Mean $C_l$', fontsize = label_fs)
            elif j == 2: ylabel(r'Std $C_d$', fontsize = label_fs)
            else: ylabel(r'Std $C_l$', fontsize = label_fs)

tight_layout()
```

```python
figure(figsize = (15,12))

figure_stats = [
    ['mean_Cd_full','mean_Cd_midEps','mean_Cd_mid','mean_Cd_close','mean_Cd_bdry'],
    ['mean_Cl_full','mean_Cl_midEps','mean_Cl_mid','mean_Cl_close','mean_Cl_bdry'],
    ['std_Cd_full','std_Cd_midEps','std_Cd_mid','std_Cd_close','std_Cd_bdry'],
    ['std_Cl_full','std_Cl_midEps','std_Cl_mid','std_Cl_close','std_Cl_bdry']
]

for j in range(4):
    for i in range(5):
        
        mean_errors = [np.mean(np.abs(np.array(basic_statistics[figure_stats[j][i]][0]) - \
                                      np.array(basic_statistics[figure_stats[j][i]][k])) / \
                       np.std(np.array(basic_statistics[figure_stats[j][i]][0]))) \
                       for k in range(1,5)]
        
        subplot(4,5,5*j+i+1)
        semilogy(np.arange(4), mean_errors, '.', markersize=10)
    
        if j == 0:
            if i == 0: title('Full domain POD', fontsize = label_fs)
            elif i == 1: title('Mid + epsilon domain POD', fontsize = label_fs)
            elif i == 2: title('Mid domain POD', fontsize = label_fs)
            elif i == 3: title('Close domain POD', fontsize = label_fs)
            else: title('Boundary POD', fontsize = label_fs)
                
        if i == 0:
            if j == 0: ylabel(r'NMAE(Mean $C_d$)', fontsize = label_fs)
            elif j == 1: ylabel(r'NMAE(Mean $C_l$)', fontsize = label_fs)
            elif j == 2: ylabel(r'NMAE(Std $C_d$)', fontsize = label_fs)
            else: ylabel(r'NMAE(Std $C_l$)', fontsize = label_fs)

tight_layout()
```

```python
transient = int(0.8*full_results['tau']/full_results['dt']) # skip first 4/5 of run
n_runs = len(full_results['Cd_true'])

Cd_keys = ['Cd_true','Cd_10','Cd_25','Cd_100','Cd_500']
Cl_keys = ['Cl_true','Cl_10','Cl_25','Cl_100','Cl_500']

EE_statistics = {}

EE_statistics['max_Cd_full'] = [[np.max(full_results[key][j][transient:]) 
                                    for j in range(n_runs)] for key in Cd_keys]
EE_statistics['max_Cd_midEps'] = [[np.max(midEps_results[key][j][transient:]) 
                                      for j in range(n_runs)] for key in Cd_keys]
EE_statistics['max_Cd_mid'] = [[np.max(mid_results[key][j][transient:]) 
                                   for j in range(n_runs)] for key in Cd_keys]
EE_statistics['max_Cd_close'] = [[np.max(close_results[key][j][transient:]) 
                                     for j in range(n_runs)] for key in Cd_keys]
EE_statistics['max_Cd_bdry'] = [[np.max(bdry_results[key][j][transient:]) 
                                     for j in range(n_runs)] for key in Cd_keys]

EE_statistics['min_Cl_full'] = [[np.min(full_results[key][j][transient:]) 
                                    for j in range(n_runs)] for key in Cl_keys]
EE_statistics['min_Cl_midEps'] = [[np.min(midEps_results[key][j][transient:]) 
                                      for j in range(n_runs)] for key in Cl_keys]
EE_statistics['min_Cl_mid'] = [[np.min(mid_results[key][j][transient:]) 
                                   for j in range(n_runs)] for key in Cl_keys]
EE_statistics['min_Cl_close'] = [[np.min(close_results[key][j][transient:]) 
                                     for j in range(n_runs)] for key in Cl_keys]
EE_statistics['min_Cl_bdry'] = [[np.min(bdry_results[key][j][transient:]) 
                                     for j in range(n_runs)] for key in Cl_keys]

EE_statistics['max_CdCl_full'] = [[np.max(full_results[key1][j][transient:]/full_results[key2][j][transient:])
                                      for j in range(n_runs)] for (key1,key2) in zip(Cd_keys,Cl_keys)]
EE_statistics['max_CdCl_midEps'] = [[np.max(midEps_results[key1][j][transient:]/midEps_results[key2][j][transient:]) 
                                        for j in range(n_runs)] for (key1,key2) in zip(Cd_keys,Cl_keys)]
EE_statistics['max_CdCl_mid'] = [[np.max(mid_results[key1][j][transient:]/mid_results[key2][j][transient:]) 
                                     for j in range(n_runs)] for (key1,key2) in zip(Cd_keys,Cl_keys)]
EE_statistics['max_CdCl_close'] = [[np.max(close_results[key1][j][transient:]/close_results[key2][j][transient:]) 
                                       for j in range(n_runs)] for (key1,key2) in zip(Cd_keys,Cl_keys)]
EE_statistics['max_CdCl_bdry'] = [[np.max(bdry_results[key1][j][transient:]/bdry_results[key2][j][transient:]) 
                                       for j in range(n_runs)] for (key1,key2) in zip(Cd_keys,Cl_keys)]
```

```python
figure(figsize = (15,9))

figure_stats = [
    ['max_Cd_full','max_Cd_midEps','max_Cd_mid','max_Cd_close','max_Cd_bdry'],
    ['min_Cl_full','min_Cl_midEps','min_Cl_mid','min_Cl_close','min_Cl_bdry'],
    ['max_CdCl_full','max_CdCl_midEps','max_CdCl_mid','max_CdCl_close','max_CdCl_bdry']
]

plot_k = [3,4]

for j in range(3):
    for i in range(5):
        subplot(3,5,5*j+i+1)
        
        qmin = np.min(np.concatenate([np.concatenate([EE_statistics[figure_stats[j][i]][0],
                                                      EE_statistics[figure_stats[j][i]][k]])
                                      for k in plot_k]))
        qmax = np.max(np.concatenate([np.concatenate([EE_statistics[figure_stats[j][i]][0],
                                                      EE_statistics[figure_stats[j][i]][k]])
                                      for k in plot_k]))
        
        for k in plot_k:
            
            if k == 0: label = 'Full'
            elif k == 1: label = 'r=10'
            elif k == 2: label = 'r=25'
            elif k == 3: label = 'r=100'
            else: label = 'r=500'
                
            plot(EE_statistics[figure_stats[j][i]][0], 
                 EE_statistics[figure_stats[j][i]][k], 
             '.', markersize=10, label=label)
            
            if i==0 and j==0: legend(loc = 'upper left', fontsize = label_fs)
            xlim([qmin - 0.05*(qmax-qmin), qmax + 0.05*(qmax-qmin)])
            ylim([qmin - 0.05*(qmax-qmin), qmax + 0.05*(qmax-qmin)])
            plot([-50,50],[-50,50],'k--', linewidth=0.5, zorder=-1)
    
        if j == 0:
            if i == 0: title('Full POD', fontsize = label_fs)
            elif i == 1: title('Mid + Eps domain POD', fontsize = label_fs)
            elif i == 2: title('Mid POD', fontsize = label_fs)
            elif i == 3: title('Close domain POD', fontsize = label_fs)
            else: title('Boundary POD', fontsize = label_fs)
                
        if i == 0:
            if j == 0: ylabel(r'Pred. Max $C_d$', fontsize = label_fs)
            elif j == 1: ylabel(r'Pred. Min $C_l$', fontsize = label_fs)
            else: ylabel(r'Pred. Max $C_d/C_l$', fontsize = label_fs)

tight_layout()
```
