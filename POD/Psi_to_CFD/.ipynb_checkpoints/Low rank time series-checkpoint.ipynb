```python
results = close_results # select one for plotting
```

```python
Cd_true = results['Cd_true']
Cl_true = results['Cl_true']
Cd_10 = results['Cd_10']
Cl_10 = results['Cl_10']
Cd_25 = results['Cd_25']
Cl_25 = results['Cl_25']
Cd_100 = results['Cd_100']
Cl_100 = results['Cl_100']
Cd_500 = results['Cd_500']
Cl_500 = results['Cl_500']

tau = results['tau']
dt = results['dt']
t = np.linspace(dt,tau,int(tau/dt))
n_samp = len(Cd_true)
```

```python
t
```

```python
figure(figsize = (15,8))
label_fs = 16

sample_nums = [2,6]

for j in range(len(sample_nums)):
    
    sample_num = sample_nums[j]
    
    subplot(len(sample_nums),2,j+1)
    plot(t,Cd_true[sample_num][1:], 'b', label='True')
    plot(t,Cd_10[sample_num][1:], 'y--', label='r=10')
    plot(t,Cd_25[sample_num][1:], 'r--', label='r=10')
    plot(t,Cd_100[sample_num][1:], 'g--', label='r=100')
    plot(t,Cd_500[sample_num][1:], 'k--', label='r=100')
    xlim(t[0],t[-1])
    ylim([0.055,0.15])
    
    title('Sample '+str(j), fontsize = label_fs)
    xticks([])
    if j == 0: 
        yticks(fontsize = label_fs)
        ylabel(r'$C_d$', fontsize = label_fs)
        legend(loc = 'upper left', ncol=3, fontsize = label_fs)
    else:
        yticks([])

    subplot(2,2,j+3)
    plot(t,Cl_true[sample_num][:-1], 'b')
    plot(t,Cl_10[sample_num][1:], 'y')
    plot(t,Cl_100[sample_num][1:], 'g')
    xlim(t[0],t[-1])
    ylim([0.45,1.1])
    
    if j == 0: 
        yticks(fontsize = label_fs)
        ylabel(r'$C_d$', fontsize = label_fs)
    else:
        yticks([])
    xticks(fontsize = label_fs)
    xlabel('Time', fontsize = label_fs)
    
tight_layout()
```
