```python
%pylab inline
import numpy as np
import numba as nb
from scipy.signal import gaussian
from scipy.ndimage import filters
```

```python
@nb.njit
def conv_smoother(x,K):
    
    y = np.zeros_like(x)
    m = len(x)
    L = len(K)
    L2 = int(floor(L/2))
    
    for i in range(m):
        k0 = i-L2
        s = 0
        for j in range(L):
            if k0+j < 0 or k0+j>=m:
                pass
            else:
                y[i] = y[i] + x[k0+j]*K[j]
                s = s + K[j]
        y[i] = y[i] / s
        
    return y
```

```python
base_dir = '../POD/Psi_to_CFD/Results/'
results = np.load(base_dir + 'results_midEps.npy',allow_pickle=True).item()
```

```python
Cd_true = results['Cd_true']
Cl_true = results['Cl_true']
Cd_32 = results['Cd_32']
Cl_32 = results['Cl_32']
Cd_64 = results['Cd_64']
Cl_64 = results['Cl_64']
Cd_128 = results['Cd_128']
Cl_128 = results['Cl_128']
Cd_256 = results['Cd_256']
Cl_256 = results['Cl_256']

tau = results['tau']
dt = results['dt']
t = np.linspace(dt,tau,int(tau/dt))
n_samp = len(Cd_true)
```

```python
f_peak = 1.4396011097007728
width_smoother = int(3/(f_peak*dt))
scale_smoother = int(0.5/(f_peak*dt))
smoother_kern = gaussian(width_smoother, scale_smoother)
smoother_kern = smoother_kern/np.sum(smoother_kern)

# Uses scipy convolve.  Reflected endpoints bias smoothed values
# q_true = [filters.convolve1d(Cd, smoother_kern,mode='reflect') for Cd in Cd_true]
# q_32 = [filters.convolve1d(Cd, smoother_kern,mode='reflect') for Cd in Cd_32]
# q_64 = [filters.convolve1d(Cd, smoother_kern,mode='reflect') for Cd in Cd_64]
# q_128 = [filters.convolve1d(Cd, smoother_kern,mode='reflect') for Cd in Cd_128]
# q_256 = [filters.convolve1d(Cd, smoother_kern,mode='reflect') for Cd in Cd_256]

# Truncate and renormalize convolution at endpoints
q_true = [conv_smoother(Cd, smoother_kern) for Cd in Cd_true]
q_32 = [conv_smoother(Cd, smoother_kern) for Cd in Cd_32]
q_64 = [conv_smoother(Cd, smoother_kern) for Cd in Cd_64]
q_128 = [conv_smoother(Cd, smoother_kern) for Cd in Cd_128]
q_256 = [conv_smoother(Cd, smoother_kern) for Cd in Cd_256]
```

```python
sample_nums = np.arange(1,10)

figure(figsize = (15,4*len(sample_nums)))
label_fs = 16

for j in range(20):
    
    sample_num = j
    
    subplot(10,2,j+1)
    plot(t,q_true[sample_num][1:], 'b', label='True')
    plot(t,q_32[sample_num][1:], 'y--', label='r=10')
    plot(t,q_64[sample_num][1:], 'r--', label='r=25')
    plot(t,q_128[sample_num][1:], 'g--', label='r=100')
    plot(t,q_256[sample_num][1:], 'k--', label='r=500')
    xlim(t[0],t[-1])
    
    yticks(fontsize = label_fs)
    ylabel(r'Sample '+str(sample_num), fontsize = label_fs)
    
    if j == 0: 
        title(r'$q$', fontsize = label_fs+2)
        legend(loc = 'upper left', ncol=3, fontsize = label_fs)
    elif j == len(sample_nums)-1:
        xticks(fontsize = label_fs)
        xlabel('Time', fontsize = label_fs)
    else:
        xticks([])

tight_layout()
```
